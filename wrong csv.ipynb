```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
game_df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
game_df.head()
```

```python
renamed_df = game_df.rename(columns={"Critic_Score": "Critic Score", "Critic_Count": "Critic Count",
                                    "User_Score": "User Score", "User_Count": "User Count", "Global_Sales": "Global Sales"})
renamed_df.head()
```

```python
reviews_df = renamed_df[["Name", "Platform","Global Sales","Critic Score", "Critic Count", "User Score", "User Count",]]
reviews_df.head()
```

```python
new_review_df = reviews_df.groupby(["Name"]).mean()
new_review_df
```

```python
reviews_df.count()
```

```python
reviews_df = reviews_df.dropna(how='any')
```

```python
reviews_df.count()
```

```python
reviews_df.head()
```

```python
reviews_df.dtypes
```

```python
reviews_df['User Score'] = reviews_df['User Score'].astype(float)
reviews_df.dtypes
```

```python
critic_reviews_df = reviews_df[["Name", "Critic Score", "Critic Count"]]
critic_reviews_df
```

```python
critic_good_reviews_df = critic_reviews_df.loc[critic_reviews_df['Critic Score'] > 70]
critic_good_reviews_df = critic_good_reviews_df.sort_values("Critic Score",ascending=False)
ten_critic_reviews_df = critic_good_reviews_df.head(20)
ten_critic_reviews_df
```

```python
high_critic_count_df = critic_good_reviews_df.sort_values("Critic Count",ascending=False)
critic_game_df = high_critic_count_df .head(20)
critic_game_df
```

```python
user_reviews_df = reviews_df[["Name","Global Sales","User Score","User Count"]]
user_reviews_df
```

```python
user_good_reviews_df = user_reviews_df.loc[user_reviews_df['User Score'] > 7]
user_good_reviews_df = user_good_reviews_df.sort_values("User Score", ascending=False)
ten_user_reviews_df = user_good_reviews_df.head(20)
ten_user_reviews_df
```

```python
high_user_count_df = user_good_reviews_df.sort_values("User Count", ascending=False)
user_game_df = high_user_count_df.head(20)
user_game_df
```

```python
critic_count_x_axis = np.arange(len(critic_game_df))
critic_count_tick_locations = [value for value in critic_count_x_axis]
```

```python
plt.figure(figsize=(20,10))
plt.bar(critic_count_x_axis, critic_game_df["Critic Score"], color='r', alpha=0.5, align="center")
plt.xticks(critic_count_tick_locations, critic_game_df["Name"], rotation="vertical")
plt.title("Critic's Top Ten Games")
plt.xlabel("Games")
plt.ylabel("Scores")
plt.show()
```

```python
user_count_x_axis = np.arange(len(user_game_df))
user_count_tick_locations = [value for value in user_count_x_axis]
```

```python
plt.figure(figsize=(20,10))
plt.bar(user_count_x_axis, user_game_df["User Score"], color='r', alpha=0.5, align="center")
plt.xticks(user_count_tick_locations, user_game_df["Name"], rotation="vertical")
plt.title("User's Top Ten Games")
plt.xlabel("Games")
plt.ylabel("Score")
plt.show()
```

```python
merge_df = pd.merge(critic_game_df, user_game_df, on="Name")
merge_df
```
